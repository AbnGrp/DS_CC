# Matplot Basics

### A simple example

```python
#Matplotlib graphs data into Figures, each of which can contain one or more Axes, an area where points can be specified 
#in terms of x-y coordinates (or theta-r in a polar plot, x-y-z in a 3D plot, etc.).

#Creating a Figure

import matplotlib as mlp
from matplotlib import pyplot as plt

fig,ax= plt.subplots() #creates a figure containing a single axes
ax.plot([1,2,3,4],[1,2,3,4])
```

### Parts of a figure

```python
#Creating figures

fig=plt.figure() #an empty figure with no axes

fig, ax= plt.subplots() #a figure with one axes

fig, ax=plt.subplots(2,2) #a figure with a 2x2 grid of axes

fig, ax=plt.subplot_mosaic(["letf","right_top"],["left","right_bottom"]) #a figure with an axes on the left and two axes on the right
```

### Coding styles

```python
#There are essentially two ways of using Matplotlib:

#Explicitly creating Figures and Axes, and call methods on them (the object-oriented style).

#Rely on pyplot to implicitly create and manage the Figures and Axes, and use pyplot functions for plotting.

#Using the OOP style

import numpy as np

x=np.linspace(0,2,100)
fig,ax=plt.subplots(figsize=(5,2.7),layout="constrained")
ax.plot(x,x,label="linear")
ax.plot(x,[i**2 for i in x],label="quadratic")
ax.plot(x,[i**3 for i in x],label="cubic")
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("Simple plot")
ax.legend()
```

```python
#Using the pyplot style

import numpy as np

x=np.linspace(0,2,100)

plt.figure(figsize=(5,2.7),layout="constrained")
plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="quadratic")
plt.plot(x,x**3,label="cubic")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple plot")
plt.legend()
```

### Styling artists

```python
#Most plotting methods have styling options

#Setting styling options


data1=[i for i in range(20)]
data2=[i for i in range(100,120)]
data3=[i for i in range(200,220)]
fig, ax=plt.subplots(figsize=(5,2.7),layout="constrained")
x=np.arange(len(data1))
ax.plot(x,np.cumsum(data1),color="blue",linewidth=3, linestyle="--")
l,=ax.plot(x,np.cumsum(data2),color="orange",linewidth=2)
l.set_linestyle(":")
```

```python
#Creating a scatter plot

fig, ax=plt.subplots(figsize=(5,2.7),layout="constrained")
ax.scatter(data1, data2, s=50, facecolor="C0",edgecolor="k")
```

### Labelling plots

```python
#The metods: set_xlabel, set_ylabel, set_title are used to add text in the indicated locations

#Adding text to a Figure

mu, sigma= 115, 15

x=mu+sigma *np.random.randn(10000)
fig,ax=plt.subplots(figsize=(5,2.7), layout="constrained")

n,bins,patches=ax.hist(x,50,density=True,facecolor="C0",alpha=0.75)

ax.set_xlabel("Length [cm]")
ax.set_ylabel("Probability")
ax.set_title("Aardvark lengths")
ax.text(75, .025, r"$\mu=115, \sigma=15 $")
ax.axis([55,175,0,0.03])
ax.grid(True)
```

```python
#Just as with lines, you can customize the properties by passing keyword arguments into the text functions

#Customizing a text function

ax.set_xlabel("my data",fontsize=14, color="red")

#Adding a TeX equation to a title

ax.set_title(r'$\sigma_i=15$')
```

```python
#We can also annotate points on a plot

fig, ax=plt.subplots(figsize=(5,2.7),layout="constrained")
t=np.arange(0.0,5.0,0.01)
s=np.cos(2*np.pi*t)
line,=ax.plot(t,s,lw=2)

ax.annotate("localmax",xy=(2,1),xytext=(3,1.5),arrowprops=dict(facecolor="black",shrink=0.05))
```

```python
#Adding a legend to a Figure

fig, ax=plt.subplots(figsize=(5,2.7),layout="constrained")
ax.plot(np.arange(len(data1)),data1,label="data1")
ax.plot(np.arange(len(data2)),data2,label="data2")
ax.plot(np.arange(len(data3)),data3, "d",label="data3")
ax.legend()
```

### Axis scales and ticks

```python
#In addition to the linear scale, Matplotlib supplies non-linear scales, such as log-scale. Since log-scales are used so much 
#there are also direct methods like loglog, semilogx and semilogy

#Plotting using log scale

fig, ax =plt.subplots(1,2,figsize=(5,2.5),layout="constrained")
xdata=np.arange(len(data1))
data=[10**i for i in data1]
ax[0].plot(xdata,data)
ax[1].set_yscale("log")
ax[1].plot(xdata,data)
```

```python
#Each Axis has a tick locator and a formatter that choose where along the Axis objects to put tick marks

fig, ax=plt.subplots(2,1,layout="constrained")
ax[0].plot(xdata,data1)
ax[0].set_title("Automatic Ticks")

ax[1].plot(xdata,data1)
ax[1].set_xticks(np.arange(0,100,30),["zero","30","sixty","90"])
ax[1].set_yticks([-1.5,0,1.5])
ax[1].set_title("Manual Ticks")
```
